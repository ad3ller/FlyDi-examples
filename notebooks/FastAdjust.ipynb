```julia
using PyPlot
using DataFrames
push!(LOAD_PATH, "C:\\Users\\Adam\\Documents\\Git\\FlyDi.jl\\src")
using FlyDi
```

```julia
fil = "../data/quad.h5"
r0 = Vector([-0.0155, -0.008, 0.0]); # coord of pa grid [1, 1] (meters)
fa = h5read_pa(fil, r0);
```

```julia
# electrodes
fig, ax = subplots(figsize=(6, 4))
ax.imshow(fa.el[101, :, :], cmap="bone_r")
;
```

```julia
# electrodes
fig, ax = subplots(figsize=(4, 6))
ax.imshow(fa.el[:, 13, :], cmap="bone_r")
;
```

```julia
voltages = [1.0, -1.0, 1.0, -1.0]
phi = potential(fa, voltages); # indexing order: z, y, x

# plotting
fig, ax = subplots(1, 1, figsize=(6, 4))
c = contourf(phi[101, :, :], 18)
colorbar(c)
;
```
